# Descriptive Statistics

```python
resistance = np.array([
    [3,9.1,9.3,9.5,9.8,9.8,10.4,10.4,10.5,10.5,11], #Dataset 1/Jaguar
    [9.1,9.1,9.1,9.2,9.2,9.9,10.1,10.8,11,20, np.nan] #Dataset 2/Panther
]).T
resistance
```

```python
df = pd.DataFrame(resistance, columns = ['Jaguar','Panther'])
df
```

```python
df.info()
```

```python
type(df)
```

```python
df.describe()
```

```python
jag_median = df['Jaguar'].median()
pan_median = df['Panther'].median()

print(f'Jaguar Median = {jag_median:.2f} ohms')
print(f'Panther Median = {pan_median:.2f} ohms')
```

```python
jag_mode = df['Jaguar'].mode()
pan_mode = df['Panther'].mode()
```

```python
jag_mode
```

```python
jag_mode.info()
```

```python
type(jag_mode)
```

```python
print('Jaguar Mode')

for i in jag_mode:
    print(i)
```

```python
print('Panther Mode')

for i in pan_mode:
    print(i)
```

```python
jQ1 = df['Jaguar'].quantile(0.25)
jQ3 = df['Jaguar'].quantile(0.75)

print(f'Jaguar IQR = {jQ3-jQ1:.2f} ohms')
```

```python
pQ1 = df['Panther'].quantile(0.25)
pQ3 = df['Panther'].quantile(0.75)

print(f'Panther IQR = {pQ3-pQ1:.2f} ohms')
```

```python
jag_mean = df['Jaguar'].mean()
pan_mean = df['Panther'].mean()

plt.figure(figsize=(7,4))
bars = plt.bar(['Jaguar', 'Panther'], [jag_mean,pan_mean],
               color="#fc43d4")
plt.title('Average Resistance by Machine')
plt.ylabel('Resistance (ohms)')

plt.bar_label(bars,padding=3)
plt.margins(y=0.1)
plt.show
```
