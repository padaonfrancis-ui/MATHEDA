# Iterating Over a List

```python
fruits = ["apple", "orange", "grapes"]

for i in fruits:
    print(i)
```

```python
for n in range(1,10):
    print(n)
```

# Iterating Over a Dictionary

```python
for char in "Python":
    print(char)
```

```python
employee = {
    'name':'ada',
    'position':'analyst',
    'salary':500000
}
```

```python
employee['position']
```

```python
for key in employee:
    print(key)
```

```python
employee = {
    'name':['ada','love'],
    'position':['analyst','engineer'],
    'salary':[80000, 11000]
}
```

```python

employee['salary']
```
